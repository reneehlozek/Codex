```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import synphot as synph
import pysynphot as S
```

```python
data=np.transpose(np.genfromtxt("../../OA44f.all1106.375360",skip_header=4))
radius = 0.1012800e15 # cm
```

```python
wave=data[0] # wavelength in micron
wave = wave*1000.0 #wavelength in nanometers
print(wave)
```

```python
lum=data[1] #luminosity in L(erg/s/um)
flux = lum/(4*np.pi*radius**2) #flux in erg/s/um/cm^2
flux = flux/10000.0 # flux in erg/s/A/cm^2
print(flux)
plt.plot(flux)
```

```python
#sp = S.ArraySpectrum(wave=wave, flux=flux, waveunits='nm', fluxunits='photlam')
synph.synphot(wave,flux, 'filter_g.dat',24.63)
```
